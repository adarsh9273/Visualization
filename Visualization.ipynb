```python
from PIL import Image

img = Image.open('content1.jpg')
img.resize((64, 64))  # resizes image in-place
plt.imshow(img)
```

```python
fig = plt.figure()

a = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(img)
a.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')

a = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(img)
imgplot.set_clim(0.0, 0.7)
a.set_title('After')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```

## Categorical Data

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

# SEABORN

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
#sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)  # alternate method
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
sns.relplot(x="time", y="value", kind="line", data=df)
#sns.lineplot(x="time", y="value", data=df)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips) # use 'row' for rows
```

```python
fmri = sns.load_dataset("fmri")

sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'frontal'"));
```

## Categorical Data

```python
sns.catplot(x="day", y="total_bill", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips);
```

```python
titanic = sns.load_dataset("titanic")

sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)  # bar plot
```

```python
sns.catplot(x="deck", kind="count", data=titanic)  # similar to histogram.
```

```python
sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=titanic);
```

```python
x = np.random.normal(size=100)
sns.distplot(x)
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])

sns.jointplot(x="x", y="y", data=df)
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris);

# alternate 
#g = sns.PairGrid(iris)
#g.map_diag(plt.hist)
#g.map_offdiag(plt.scatter);
```
